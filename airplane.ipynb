```python
days = {0:'Monday', 1:'Tuesday', 2:'Wednesday', 3:'Thursday', 4:'Friday', 5:'Saturday', 6:'Sunday'}
```

```python
airplane_df['DayOfWeek'] = airplane_df['DayOfWeek'].map(days)
airplane_df
```

```python
## Create Features DataFrame
features_df = airplane_df.filter(['Flight','Airline', 'AirportFrom', 'AirportTo','DayOfWeek','Time','Length' ], axis = 1)
features_df
```

```python
#create target df
target_df = airplane_df.filter(['Flight','Airline','Delay'], axis=1)
target_df
```

```python
features_df = features_df.rename(columns={'Airline': 'airline', 'AirportFrom' : 'departing_airport','AirportTo': 'arrival_airport','DayOfWeek' : 'weekday','Time' : 'delay_mins','Length' : 'flight_length'})
features_df
```

```python
# Add this import for sql database
import sqlite3
```

```python
# connecting to 'features.db'
conn = sqlite3.connect('Airlines.db')
```

```python
# Adding feature_df database
features_df.to_sql(name ='features', con=conn, if_exists='replace', index=False)
conn.commit
```

```python
# Adding target_df to database
target_df.to_sql(name ='target', con=conn, if_exists='replace', index=False)
conn.commit
```
